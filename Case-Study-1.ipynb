# Polynomial Regression

```python
X=df.drop(['cnt'],axis=1)
y=df['cnt']
X=X.drop(['workingday','atemp'],axis=1)
```

```python
from sklearn.preprocessing import PolynomialFeatures

pf=PolynomialFeatures()

x=pf.fit_transform(X)

feature_names=pf.get_feature_names(X.columns)
x=pd.DataFrame(x,columns=feature_names)
```

```python
x.info(verbose=True)
```

```python
model=sm.OLS(y,x)
result=model.fit()
result.summary()
```

# backward elimination

```python
cols = list(x.columns)
while (len(cols)>0):
    X_1 = x[cols]
    model = sm.OLS(y,X_1).fit()
    p = pd.Series(model.pvalues.values,index = cols)      
    pmax = max(p)
    feature_maxp = p.idxmax()
    if(pmax>0.05):
        cols.remove(feature_maxp)
    else:
        break
selected_features_BE = cols
print(selected_features_BE)
        
```

```python
x=x.loc[:,selected_features_BE]


model=sm.OLS(y,x)
result=model.fit()
result.summary()
```

# Linear Regression-Machine Learning

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score


df=pd.read_csv('bike sharing by day.csv')
df.head(5)
```

```python
data=df.drop(['dteday','instant','casual','registered','atemp','workingday'],axis=1)
data.head(5)
```

```python
cat_cols=['season','mnth','weekday','weathersit']
data=pd.get_dummies(data,columns=cat_cols,drop_first=True)

data['yr']=data['yr'].map({0:2011,1:2012})
```

```python
x=data.drop(['cnt'],axis=1)
y=data['cnt']
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=45)
```

```python
lr=LinearRegression()
lr.fit(X_train,y_train)
y_train_pred=lr.predict(X_train)
print('R2 of train data',r2_score(y_train,y_train_pred))


y_test_pred=lr.predict(X_test)
print('R2 of test data',r2_score(y_test,y_test_pred))
```

```python
cf=pd.DataFrame([X_train.columns,lr.coef_]).T
cf.columns=['cols','coefs']
cf
```

# Feature Selection using RFE

```python
from sklearn.feature_selection import RFE
```

```python
model=LinearRegression()
rfe=RFE(model,10)

X_rfe=rfe.fit_transform(X_train,y_train)
model.fit(X_rfe,y_train)


print(rfe.support_)
print(rfe.ranking_)
```

```python
pd.DataFrame([X_train.columns,rfe.ranking_]).T
```

```python
no_of_cols=len(X_train.columns)

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=45)


score_list=[]

for n in range(no_of_cols):
    model=LinearRegression()
    rfe=RFE(model,n+1)
    X_train_rfe=rfe.fit_transform(X_train,y_train)
    model.fit(X_train_rfe,y_train)
    
    X_test_rfe=rfe.transform(X_test)
    score=model.score(X_test_rfe,y_test)
    
    score_list.append(score)
```

```python
plt.plot(score_list)
plt.show()
```

```python
score_list
```

```python
#we found thata r2 is maximum when no of columns are 10.Henece we run RFE with 10 to identify the best 10 features
model=LinearRegression()
rfe=RFE(model,10)

X_rfe=rfe.fit_transform(X_train,y_train)
model.fit(X_rfe,y_train)


print(rfe.support_)
print(rfe.ranking_)
```

```python
sel=pd.DataFrame([X_train.columns,rfe.ranking_]).T
sel.columns=['cols','select']
sel=sel[sel['select']==1]
sel
```

```python
selected_cols=sel['cols'].tolist()
selected_cols
```

```python
lr=LinearRegression()
lr.fit(X_train.loc[:,selected_cols],y_train)
y_train_pred=lr.predict(X_train.loc[:,selected_cols])
print('R2 of train data',r2_score(y_train,y_train_pred))


y_test_pred=lr.predict(X_test.loc[:,selected_cols])
print('R2 of test data',r2_score(y_test,y_test_pred))
```
